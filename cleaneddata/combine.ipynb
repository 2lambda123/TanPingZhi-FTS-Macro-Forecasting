```python
import numpy as np
ffr = np.load("ffr.npy", allow_pickle = True)
fyty = np.load("fyty.npy", allow_pickle = True)
re = np.load("RE_reformat.npy", allow_pickle=True)
sp500 = np.load("snp500.npy", allow_pickle = True)[:, [0, 2]]

usdx = np.load("usdx.npy", allow_pickle=True)
vix = np.load("vixclose.npy", allow_pickle = True)
wtinew = np.load("wti_reformat.npy", allow_pickle = True)
```

```python
import pandas as pd
# ffr.columns = fyty.columns = re.columns = sp500.columns = usdx.columns = vix.columns = wtinew.columns = ['t', 'v']
data_list = [sp500, ffr, fyty, re, usdx, vix, wtinew]
pdL = list(map(lambda x: pd.DataFrame(x), data_list))
for d in pdL:
    d.columns = ['t','v']

x = pdL[0].merge(pdL[1], on='t', how='outer').\
        merge(pdL[2], on='t', how='outer').\
        merge(pdL[3], on='t', how='outer').\
        merge(pdL[4], on='t', how='outer').\
        merge(pdL[5], on='t', how='outer').\
        merge(pdL[6], on='t', how='outer').\
        sort_values('t')
x.columns = ['date', "sp500", "ffr", "fyty", "re", "usdx", "vix", "wti"]
x.to_csv("combined_data.csv")
    
```
